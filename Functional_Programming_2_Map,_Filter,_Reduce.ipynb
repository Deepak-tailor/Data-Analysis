```python
def summation(x, y, *args):
    print(x)
    print(y)
    print(args)
    result = x + y
    if args:
        result = result + sum(args)

    return result
```

```python
summation(5,6,7,8,9)
```

```python
def create_person(name,age,gender,**extra_info):
  print("Name",name)
  print("age",age)
  print("gender",gender)
  print(extra_info)

  Person = {
      "name":name,
      "age":age,
      "gender":gender
  }

  if extra_info:
    Person.update(extra_info)

  return Person
```

```python
create_person(name = "Rohit", age = 26, gender = "Male",Salary = 17000,Department = 'Sales')
```
