## Concatenation and Merging

```python
# users -> stores the user details - ID and Name of the users
```

```python
users = pd.DataFrame({'user_id' : [1, 2, 3],
                      'name' : ['Sharadh', 'Rahul', 'Khushali']})
```

```python
users
```

```python
# msgs -> Stores the messages user have sent - IDs and messages
```

```python
msgs = pd.DataFrame({'user_id' : [1, 1, 2, 4],
                     'msg' : ['Hmm', 'achha', 'theek hai', 'nice']})
```

```python
# We can combine these dataframes to form a single dataframe
```

```python
pd.concat([users, msgs])
```

```python
pd.concat([users, msgs], axis = 1)
```

## Merging DataFrames

```python
users
```

```python
msgs
```

```python
users.merge(msgs, on = 'user_id')
```

```python
# Get all users
```

```python
users.merge(msgs, on = 'user_id', how = 'left')
```

```python
pd.merge(users, msgs, on = 'user_id', how = 'left')
```

```python
users.merge(msgs, on = 'user_id', how = 'right')
```
