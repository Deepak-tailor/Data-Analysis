```python
def pretty(a):
  def inner():
    print("-"*50)
    a()
    print("-"*50)

  return inner
```

```python
def calendar_update():
  print("Have you updated your calendar?")
```

```python
# For the decorator we use @sign

@pretty
def calendar_update():
  print("Have you updated your calendar?")
```

```python
@pretty
def soo():
    print("This is amazing!!!")
```

```python
pretty(message)
```

```python
pretty(calendar_update)
```
