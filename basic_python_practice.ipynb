```python
entries = int(input("How many entries you want to provide: "))

bookings = {}

print(f"Enter the name and amount for {entries} entries: ")

for _ in range(entries):
    name = input("Name: ")
    amount = float(input("Amount: "))
    if name in bookings:
      bookings[name]= bookings[name]+ amount
    else:
      bookings[name] = amount

sorted_booking = dict(sorted(bookings.items(), key=lambda x: x[0], reverse=True))
print("Sorted Bookings",sorted_booking)
```
