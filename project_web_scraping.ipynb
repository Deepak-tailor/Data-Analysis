```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def book_scraper():
    print("Scraping Books Catalogue...")

    url = "http://books.toscrape.com/index.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "lxml")

    # 1. Find Containers
    books = soup.find_all('article', class_='product_pod')

    if not books:
        print("No books found.")
        return None

    # Initialize Lists
    titles_list = []
    prices_list = []
    stock_list = []
    ratings_list = []

    # 2. Loop through containers
    for book in books:
        # --- TITLE ---
        try:
            title = book.find('h3').find('a')['title']
        except:
            title = "N/A"
        titles_list.append(title)

        # --- PRICE ---
        try:
            price = book.find('p', class_='price_color').text.strip()
        except:
            price = "N/A"
        prices_list.append(price)

        # --- STOCK ---
        try:
            stock = book.find('p', class_='instock availability').text.strip()
        except:
            stock = "N/A"
        stock_list.append(stock)

        # --- RATING ---
        try:
            # Get the second class name (e.g., 'Three', 'Four', 'Five')
            rating = book.find('p', class_='star-rating')['class'][1]
        except:
            rating = "N/A"
        ratings_list.append(rating)

    # 3. Create Dictionary
    data = {
        "Book Title": titles_list,
        "Price": prices_list,
        "Availability": stock_list,
        "Rating": ratings_list
    }

    print(f"Successfully scraped {len(titles_list)} books.")
    return data
```

```python
book_data = book_scraper()
```
