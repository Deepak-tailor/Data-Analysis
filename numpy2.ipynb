Logical Operators


*   any
*   all



```python
# any : will give True in O/P even if single value satisfies the condition
# all : will give True in O/P if every value satisfied the condition
```

```python
# any -> or
# all -> and
```

```python
# heights = 176, 172, 162, 154, 166

# a => 175 greater than any of the student => Yes
# a => 175 greater than all of the student -> No
```

```python
# Price of items in your shopping list
```

```python
Prices = np.array([50, 45, 25, 20, 35])

# Your budget
budget = 30
# check if there's atleast one item that you can afford?
can_afford = (Prices <= budget).any()

if can_afford:
  print("You can afford atleast one item")
else:
  print("You can't afford atleast one item")
```

```python
# np.where
```

```python
# product prices
prices = np.array([45, 55, 60, 75, 40, 90])

# Apply a 10% discount on product having prices above $50
discounted_prices = np.where(prices > 50, prices * 0.9, prices)

print(discounted_prices)
```
